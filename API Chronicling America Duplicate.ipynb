```python
#now we will create a empty list that will contain data 
data = []
```

```python
# Setting parameters (every term is given a value)
start_date = '1920'
end_date = '1952'
search_term = 'Immigration'
state = 'New York'
```

```python
for k in range(1, total_pages+1): 
    url = (f'https://chroniclingamerica.loc.gov/search/pages/results/?state={state}&date1={start_date}'
           f'&date2={end_date}&proxtext={search_term}&x=11&y=7&dateFilterType=yearRange&rows=20'
           f'&searchType=basic&format=json&page={k}')
#requesting url 
    information = requests.get(url)
#making sure info is raw
    raw = information.text
#printing status code of each page, making sure they all work
    print(f'page {k} status code:', information.status_code)
    answer= json.loads(raww)
    items_ = answer['items']
    

    for item_ in items_:
        row_datas= {}
        try: 
         row_datas['title'] = item_['title_normal']
        except:
          row_datas['title'] = "none"
        try:
          row_datas['city'] = item_['city']
        except:
          row_datas['city'] = "none"
        try:
          row_datas['date'] = item_['date']
        except:
          row_datas['date'] = "none"
        try:
          row_datas['raw_text'] = item_['ocr_eng']
        except:
          row_datas['raw_text'] = 'none'
        data.append(row_datas)
```

```python
# putting data into data frame 
df = pd.DataFrame.from_dict(data)
```

```python
# uploading spacy text 
nlp= spacy.load('en_core_web_sm')

def process_text (text): 
    """We are removing new lines and sorting the text"""
    text = text.replace ('\n', '')
    
    doc = nlp(text)
    
    #extracting the tokens and giving them variable "tokens"
    tokens =[token for token in doc]
    
    #labeling the tokens "token" if not it will be "stop word"
    no__stops = [token for token in tokens if not token.is_stop]
    
    #labeling the tokens "token" if not it will be an alphabetic character" 
    no_punct = [ token for token in no__stops if not token.is_alpha]
    
    #searching for lemmas in no_punct
    lemmas = [token.lemma_ for token in no_punct]
    
    #putting lemma in lemmas lowercase
    lemmas_lower = [lemma.lower() for lemma in lemmas ]
    
    #making lemmas in string
    lemmas_string = ''.join(lemmas_lower)
    return lemmas_string
```

```python
# make dataframe lemmas the total of 
df['lemmas']= df['raw_text'].apply(process_text)
```
